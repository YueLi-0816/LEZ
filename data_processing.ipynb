```python
import pandas as pd
```

### Calculate the annual flow of each site

```python
def get_annual_flow(site, year):
    
    data1 = pd.read_csv('../Data/Data processing/20211028_SCOOT_Data/JupyterNotebook_CSV/20211213_SCOOT_data/'+ site +'&2019_08_05-2021_12_12.csv')
    data2 = pd.read_csv('../Data/Data processing/Traffic data/SCOOT/data/'+ site +'&2021_12_13-2023_09_30.csv')
    data = pd.concat([data1, data2]).reset_index(drop=True)
    
    data['year'] = data['lastUpdate'].str.split('-').str[0]
    all_flow = data.loc[data.year == str(year), 'flow'].sum()
    flow_annual = pd.DataFrame({'year': year, 'siteId': site, 'annual_flow': all_flow, 'description': data['originDescription'][0],
                               'lat': data['originLat'][0], 'lon': data['originLong'], 'easting': data['originEasting'][0], 
                               'northing': data['originNorthing'][0]}, index=[0])
                        
    return flow_annual
```

```python
get_annual_flow('GA0401_R',2021)
```

### Get the annual flow of 530 sites in 2021

```python
site_df = pd.read_csv('../Data/Data processing/20211028_SCOOT_Data/JupyterNotebook_CSV/useful_detector_530.csv')
```

```python
flow_list = []
for site in site_df.siteId:
    annual_flow = get_annual_flow(site,2021)
    flow_list.append(annual_flow)
    
annual_flow_530 = pd.concat(flow_list).reset_index(drop=True)
annual_flow_530
```
